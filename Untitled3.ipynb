```python
import numpy as np
import pandas as pd

print("am using mine dataset")
df = pd.read_csv("mine.csv")
```

```python
#
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.info())
```

```python
#return the new data Frame with no empty cells
new_df = df.dropna()
print(new_df.to_string())
```

```python
#to remove all null values from the origin dataset
df.dropna(inplace=True)
print(df.to_string())
```

```python
df.dropna(inplace=False)
print(df.to_string())
```

```python
#allows us replace an empty cell with a value.its not specific
df.fillna(50,inplace=True)
```

```python
#repalce null values in a specific column
df["Calories"].fillna(90,inplace=True)
```
